```python
# Assuming two_aux['current_R'] is a 3D numpy array
current_R = two_aux['current_R']

# Create a figure with subplots for each layer
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

for i in range(3):  # Loop through layers 0, 1, 2
    im = axes[i].imshow(current_R[:, :, i])  # Plot the i-th layer
    axes[i].set_title(f"Layer {i}")  # Set the title for the subplot
    cbar = fig.colorbar(im, ax=axes[i])  # Add a colorbar to the subplot

# Add a global title to the figure
fig.suptitle("Layers of current_R")

# Show the plot
plt.tight_layout()
plt.show()
```
